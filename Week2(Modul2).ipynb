Pertanyaan
1. Apakah perbedaan gambar yang ditampilkan tanpa dan dengan matplotlib? Gambar yang ditampilkan tanpa Matplotlib cenderung terlihat sederhana dan kurang diatur, sehingga sulit untuk dimodifikasi atau diberi elemen tambahan seperti label atau skala warna. Sebaliknya, dengan Matplotlib, gambar bisa diberi pengaturan tambahan seperti sumbu, label, atau skala warna yang membantu menyampaikan informasi dengan lebih jelas dan mudah dipahami.
2. Apakah perbedaan dan pengaruhnya pembuatan black image antara tipe data int16
dan int32? int16 memiliki rentang nilai yang cukup kecil, sehingga cocok digunakan untuk gambar dengan resolusi rendah atau yang tidak memerlukan detail warna yang kompleks. Sementara itu, tipe data int32, yang memiliki rentang hingga 2,14 miliar, lebih ideal digunakan untuk gambar beresolusi tinggi atau yang memerlukan banyak detail, karena bisa menyimpan nilai yang lebih besar tanpa menyebabkan overflow saat melakukan operasi pada nilai piksel.
3. Apakah kegunaan google.colab.patches import cv2_imshow pada potongan kode
berikut. Fungsi ini digunakan untuk menampilkan gambar yang diproses menggunakan OpenCV, sebuah pustaka yang sering digunakan dalam pengolahan gambar dan video
4. Apakah kegunaan skimage import io pada potongan kode soal nomor 3. Fungsi ini digunakan untuk memuat gambar dari file, misalnya dengan membaca format seperti JPEG, PNG, atau TIFF, dan menampilkannya secara langsung.


```python
kitten = cv.imread('/content/drive/MyDrive/download.jpg')

blue, green, red = cv.split(kitten)

rb_kitten = cv.merge([blue, green*0, red])
gb_kitten = cv.merge([blue, green, red*0])

fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].imshow(cv.cvtColor(rb_kitten, cv.COLOR_BGR2RGB))
axes[0].set_title('Channel Red-Blue')

axes[1].imshow(cv.cvtColor(gb_kitten, cv.COLOR_BGR2RGB))
axes[1].set_title('Channel Green-Blue')

plt.show()
```
